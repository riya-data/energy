Cleaning the dataset that might entail various significant inconsistencies.

# Dataframe Creation

The column country has country names and continent names which was indentified while checking for unqiue values. The column 'iso_code' has some missing values and that is because they entail the alpha-3 country code for the respective country names from country column. To seperate these inconsistent values, new dataframes are created.

#### Data - Countries: 2000-2021

```python
# Creating a DataFrame for countries data from 2000-2021

e_country =  energy.query( "year >= 2000 and iso_code.notnull()")
e_country.head()
```

#### Data - Continents: 2000-2021

```python
# Creating a DataFrame for continent data from year 2000-2021 

e_continent = energy[(energy["year"]>=2000)&((energy["iso_code"].isna())|(energy["iso_code"]=="ATA"))]
e_continent = e_continent[(e_continent["country"]=="Africa")|
                          (e_continent["country"]=="Asia")|
                          (e_continent["country"]=="North America")|
                          (e_continent["country"]=="South America")|
                          (e_continent["country"]=="Antarctica")|
                          (e_continent["country"]=="Europe")|
                          (e_continent["country"]=="Oceania")]

# Checking unique values for continents

e_continent["country"].value_counts()
```

```python
# renaming country column to continent

e_continent.rename(columns={'country': 'continent'}, inplace=True)
e_continent
```

#### Data - World: 2000-2021

```python
# Creating a DataFrame for continent data from year 2000-2021 

e_world=energy.query( "country=='World'")
e_world = e_world[e_world["year"]>=2000]
e_world.head()
```

# Data Analysis

## Population across the world

```python
# Plotting population from 2000-2021

pop_dt = e_country[e_country['country']!='World'] 

global_pop = pop_dt[['country', 'year', 'population']].groupby(['country','year']).sum().reset_index()

px.choropleth(data_frame=global_pop, locations="country", locationmode="country names", animation_frame="year",
              color="population", color_continuous_scale="Aggrnyl", title="Global Population from 2000 - 2021")
```

The chart above illustrates how world population has changed throughout history. It is evident that world population has being growing at a tremendous rate and has now reached upto 8 billion.

#### Top countries with largest population as of 2021

```python
# Top countries as of 2021
ordered_pop = e_country[e_country['year']==2021].sort_values(by='population',ascending=False)

# Years
years = pop_dt['year'].unique()

# List of top countries with largest population in 2021 
top = ordered_pop['country'].head(15).unique()

list = []
for year in years:
    for cont in top:
        amount = pop_dt[pop_dt['country']==cont][pop_dt['year']==year]['population'].sum()
        list.extend([[cont, year, amount]])

global_pop = pd.DataFrame(list, columns=['country', 'year', 'population'])
plt.figure(figsize=(20,10))

fig = px.line(global_pop, x='year', y='population', color='country', 
        line_group='country', title='Countries with largest Population as of 2021')
fig.show()
```

China and India have clearly had the largest populations throughout the entire time period. This is followed by USA, Indonesia, Pakistan, Brazil and Nigeria.

#### Relationship between population vs ....

```python
with plt.rc_context(rc = {'figure.dpi': 300, 'axes.labelsize': 9,
                          'xtick.labelsize': 9, 'ytick.labelsize': 9,
                          'legend.title_fontsize': 7, 'axes.titlesize': 9, 
                          'axes.titlepad': 7}):

    p_data = e_world[e_world['country']=='World']
      
    fig_4, ax_4 = plt.subplots(2, 2, figsize = (10, 6), gridspec_kw = {'width_ratios': [4, 4], 
                                                'height_ratios': [3, 3]})
    ax_flat = ax_4.flatten()
    
    ## 1st graph
    sns.lineplot(ax=ax_flat[0], data = p_data,
             x='year', y='population', lw=2, color='navy').set_title('Yearly Global Population')
    

    ## 2nd graph
    sns.lineplot(ax=ax_flat[1], data = p_data,
                x='electricity_demand', y='population', lw=2, 
                color='navy').set_title('Population vs Global electricity demand')
       
    ## 3rd graph
    sns.lineplot(ax=ax_flat[2], data = p_data,
                x='fossil_electricity', y='population',  
                lw=2,color='navy').set_title('Population vs Global fossil fuel generation')
   
    
    ## 4th graph
    sns.lineplot(ax=ax_flat[3], data = p_data,
                x='greenhouse_gas_emissions', y='population', 
                lw=2,color='navy').set_title('Population vs Global greenhouse gas emissions')  
    
    plt.tight_layout(pad=0.4, w_pad=3, h_pad=3)
    plt.show()
```

Human population growth is a major contributor to global warming, given that humans use fossil fuels to power their increasingly mechanized lifestyles. More people means more demand for oil, gas, coal and other fuels mined or drilled from below the Earth’s surface that, when burned, spew enough carbon dioxide (CO2) into the atmosphere to trap warm air inside like a greenhouse.

These line graph show a direct relationship between population and greenhouse gas emissions - increase in popluation has no doubt been a reason in emitting greenhouse gas. From the graphs it is evident that population size and growth rates are significant factors in magnifying the impacts of global warming - increase in electricity generation using fossil fuels, therby emitting CO2 and greenhouse gas emissions.

```python
#Correlations

corr_data = e_world[['population','gdp','greenhouse_gas_emissions','per_capita_electricity',
                     'primary_energy_consumption','fossil_electricity']]
plt.figure(figsize=(10,8))
sns.heatmap(data=corr_data.corr(), annot=True, fmt='.1f',linewidth=.5, cmap="crest")
plt.show()
```

Correlation heatmaps are a type of plot that visualize the strength of relationships between numerical variables. Correlation plots are used to understand which variables are related to each other and the strength of this relationship.

From the heatmap above, there seems to be a positive correlation with global population - primary energy demand, per capita electricity generation and greenhouse gas emissions. 

Majority of developed nations have had a farily steady population throughout the time period, developing nations on the other hand have continued to see a large rise in population growth. Therefore, as the population continues to increase for the time being, which in turn is responsible for high electricity demand, causing high usage of fossil fuels.

## World Energy Distribution 

```python
# Primary energy consumption over 20 years span worldwide

#Oil Primary energy consumption
sns.lineplot(x = e_world['year'],y = e_world['oil_consumption'],
             label= 'Oil', color='navy',data=e_world, ci=None)

#Coal Primary energy consumption
sns.lineplot(x=e_world['year'],y=e_world['coal_consumption'],
             label= 'Coal',color='darkgreen',data=e_world,ci=None)

#Gas Primary energy consumption
sns.lineplot(x=e_world['year'],y=e_world['gas_consumption'],
             label= 'Gas',color='darkgoldenrod',data=e_world,ci=None)

#Hydropower Primary energy consumption
sns.lineplot(x=e_world['year'],y=e_world['hydro_consumption'],
             label= 'Hydropower',color='deepskyblue',data=e_world,ci=None)

#Nuclear Primary energy consumption
sns.lineplot(x=e_world['year'],y=e_world['nuclear_consumption'],
             label= 'Nuclear',color='purple',data=e_world,ci=None)

#Wind Primary energy consumption
sns.lineplot(x=e_world['year'],y=e_world['wind_consumption'],
             label= 'Wind',color='deeppink',data=e_world,ci=None)

#Solar Primary energy consumption
sns.lineplot(x=e_world['year'],y=e_world['solar_consumption'],
             label= 'Solar',color='orange',data=e_world,ci=None)

#Biofuels Primary energy consumption
sns.lineplot(x=e_world['year'],y=e_world['biofuel_consumption'],
             label= 'Biofuels',color='magenta',data=e_world,ci=None)

#Other renewables Primary energy consumption
sns.lineplot(x=e_world['year'],y=e_world['other_renewable_consumption'],
             label= 'Other renewables',color='maroon',data=e_world,ci=None)

plt.title("Primary Energy Consumption from various resources, World", fontsize = 15)
plt.xlabel("Year",fontsize = 12)
plt.ylabel("Terawatt-hours(TWh)", fontsize = 12)
plt.xticks(e_world['year'])
plt.rcParams["figure.figsize"] = [17, 7]
plt.rcParams["figure.autolayout"] = True
plt.tick_params(labelsize=10)
plt.show()
```

Primary energy consumption measures the total energy demand of a country. Primary energy is calculated based on the 'substitution method' which takes account of the inefficiencies in fossil fuel production by converting non-fossil energy into the energy inputs required if they had the same conversion losses as fossil fuels.

The lineplot describes primary energy demand by various energy sources over the span of 20 years.

Oil, coal, and gas, all of which are fossil fuels - primary energy demand from these fuels remain high as of 2021. However, their demand had dropped during 2019-2020, covid being the plausible reason. Primary energy consumption from hydropower is largest amongst all other renewable energy resources. This is followed by nuclear power and wind. 

There is a significant drop in the primary energy consumption from biofuels, a core reason could be the drawbacks associated to certain types of biofuels.  
For example: Biodiesel, which is usually made from soybeans, could replace conventional diesel in areas like aviation or long-haul trucking, but, for now, producing it at scale is cost-prohibitive. The feedstock inputs required for biodiesel are more expensive than petroleum.

## Carbon Intensity
#### Carbon intensity is the electricity producted, measured in grams of carbon dioxide emitted per kilowatt-hour (KWh)

```python
#Plotting carbon intensity across all continents over 20 years

with plt.rc_context(rc = {'figure.dpi': 200, 'axes.labelsize': 12,
                          'xtick.labelsize': 12, 'ytick.labelsize': 15,
                          'legend.title_fontsize': 15, 'axes.titlesize': 15,
                          'axes.titlepad': 15}):

    c_data1 = e_continent[e_continent['continent']=='Africa']
    c_data2 = e_continent[e_continent['continent']=='Asia']
    c_data3 = e_continent[e_continent['continent']=='Europe']
    c_data4 = e_continent[e_continent['continent']=='North America']
    c_data5 = e_continent[e_continent['continent']=='South America']
    c_data6 = e_continent[e_continent['continent']=='Oceania']
      
    sns.lineplot(data = c_data1, x='year', y='carbon_intensity_elec', 
             lw=2, color='navy', label='Africa')

    sns.lineplot(data = c_data2, x='year', y='carbon_intensity_elec', 
             lw=2, color='darkgreen',label='Asia')
       
    sns.lineplot(data = c_data3, x='year', y='carbon_intensity_elec', 
             lw=2, color='red',label= 'Europe')

    sns.lineplot(data = c_data4, x='year', y='carbon_intensity_elec', 
             lw=2, color='magenta', label = 'North America')

    sns.lineplot(data = c_data5, x='year', y='carbon_intensity_elec', 
             lw=2, color='purple', label = 'South America')

    sns.lineplot(data = c_data6, x='year', y='carbon_intensity_elec', 
             lw=2, color='orange', label = 'Oceania')
    
    plt.title("CO2 Emissions per KWh - over 20 years")
    plt.xlabel("Year")
    plt.ylabel("Kilowatt-hours(KWh)")
    plt.show()
```

Asia is by far the largest CO2 emitter, accounting for more than half of global emissions. China is, by a significant margin, Asia’s and the world’s largest emitter. For future research, fossil fuel consumption and production by countries will be analysed to get more detailed results.

CO2 emissions from Africa have majorly dropped and a theory would be that they have transitioned to low-carbon energy resources. This will be further analysed to prove this theory.

Carbon dioxide concentration in the atmosphere continues to rise. Being a greenhouse gas, carbon dioxide has contributed to the rise of global temperatures.

```python
# highest CO2 emissions over top 20 countries in 2021 - Carbon Intensity 
co_top = e_country[e_country['country']!='World']

# temporary dataset with carbon_intensity_elec ordered from largest to smallest
ordered_co = co_top[co_top['year']==2021].sort_values(by='carbon_intensity_elec',ascending=False)

# Years
years = co_top['year'].unique()
    
# List of top countries with largest carbon_intensity_elec 
top_countries = ordered_co['country'].head(20).unique()
    
c_list = []
for year in years:
    for cont in top_countries:
        amount = co_top[co_top['country']==cont][co_top['year']==year]['carbon_intensity_elec'].sum()
        c_list.extend([[cont, year, amount]])

# Temporary dataframe
co_df = pd.DataFrame(c_list, columns=['country', 'year', 'carbon_intensity_elec'])


fig = px.line(co_df, x='year', y='carbon_intensity_elec', color='country',  
        line_group='country', title='Top 20 Countries with highest Carbon Intensity as of 2021')
fig.show()
```

The plot shows the top 20 countries that emit highest carbon intensity of electricity production, measured in grams of carbon dioxide emitted per kilowatt-hour while producing electricity.

Estonia, Poland, and Mongolia are top countries emitting highest carbon intensity as of 2021.

Interestingly, Moldova was placed at the top in 2010 for emitting the highest carbon intensity while producing electricity, but is placed 6th in 2021. There has been major CO2 emission fluctuations while producing electricity over the years. 

## Low Carbon

Renewable power can significantly reduce global CO2 emissions. The focus must be on the decarbonision of the global energy system as it accounts for almost two-thirds of greenhouse gas emissions.
A solution to reducing CO2 emissions is by using low carbon energy resources like renewables and nuclear power.

```python
#Plotting electricity generated by Low carbon sources across all continents over 20 years

with plt.rc_context(rc = {'figure.dpi': 200, 'axes.labelsize': 12,
                          'xtick.labelsize': 12, 'ytick.labelsize': 15,
                          'legend.title_fontsize': 15, 'axes.titlesize': 15,
                          'axes.titlepad': 15}):

    c_data1 = e_continent[e_continent['continent']=='Africa']
    c_data2 = e_continent[e_continent['continent']=='Asia']
    c_data3 = e_continent[e_continent['continent']=='Europe']
    c_data4 = e_continent[e_continent['continent']=='North America']
    c_data5 = e_continent[e_continent['continent']=='South America']
    c_data6 = e_continent[e_continent['continent']=='Oceania']
      
    sns.lineplot(data = c_data1, x='year', y='low_carbon_electricity', 
             lw=2, color='navy', label='Africa')

    sns.lineplot(data = c_data2, x='year', y='low_carbon_electricity', 
             lw=2, color='darkgreen',label='Asia')
       
    sns.lineplot(data = c_data3, x='year', y='low_carbon_electricity', 
             lw=2, color='red',label= 'Europe')

    sns.lineplot(data = c_data4, x='year', y='low_carbon_electricity', 
             lw=2, color='magenta', label = 'North America')

    sns.lineplot(data = c_data5, x='year', y='low_carbon_electricity', 
             lw=2, color='purple', label = 'South America')

    sns.lineplot(data = c_data6, x='year', y='low_carbon_electricity', 
             lw=2, color='orange', label = 'Oceania')
    
    plt.title("Electricity generated from low-carbon sources(KWh)")
    plt.xlabel("Year")
    plt.ylabel("Terawatt-hours(TWh)")
    plt.show()
```

The linegrapgh represents electricity generated from low-carbon sources, measured in terawatt-hours. It is the sum of electricity generated from renewables and nuclear power. 

Positively, since 2010 Asia has managed to increase electricity generation every year using low carbon energy sources. However, countries in Africa and Oceania need to move towards using renewable power plants to help combat surge in CO2 emissions. For future analysis, various renewable sources used by Asia to generate electricity could be compared in detail.
